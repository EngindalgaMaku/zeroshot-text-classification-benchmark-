## 3. Son 2 Deneyi Çalıştır

```python
import subprocess
import os

# Son 2 deney (max_seq_length=512 ile)
experiments = [
    "exp_yahoo_answers_topics_qwen3",      # Küçük, önce bu
    "exp_SetFit_20_newsgroups_qwen3",      # Büyük, sonra bu
]

results = []

for i, exp_name in enumerate(experiments, 1):
    print(f"\n{'='*70}")
    print(f"[{i}/{len(experiments)}] Running: {exp_name}")
    print(f"{'='*70}\n")
    
    config_path = f"experiments/{exp_name}.yaml"
    
    try:
        # GPU memory temizle
        torch.cuda.empty_cache()
        
        # Deneyi çalıştır
        cmd = ["python", "main.py", "--config", config_path]
        subprocess.run(cmd, check=True)
        
        results.append((exp_name, "✅ Success"))
        print(f"\n✅ {exp_name} completed successfully!")
        
    except subprocess.CalledProcessError as e:
        results.append((exp_name, "❌ Failed"))
        print(f"\n❌ {exp_name} failed!")
        
    except Exception as e:
        results.append((exp_name, f"❌ Error: {str(e)}"))
        print(f"\n❌ {exp_name} error: {e}")

# Özet
print("\n" + "="*70)
print("SUMMARY")
print("="*70)
for exp, status in results:
    print(f"{status} {exp}")

print("\n" + "="*70)
print("🎉 ALL 63/63 EXPERIMENTS COMPLETE!")
print("="*70)
```
